# Double plot


To perform a scatter plot on top of base image/plot 


Import packages:
    - pandas for importing data for scatter plot
    - matplotlib for creating scatter plot
    - cv2 is computer vision package to import base image of Wuhan city

```python
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

Now we import the Wuhan city map as png format using matploylib python package

The imported image is then printed using plt.show()

```python
im = plt.imread('Wuhan_in_Hubei.png')
implot = plt.imshow(im)

plt.show()
```

To know the shape of the imag ethat we have imported we can leverage the .shape() paramter. This output a tuple with number of rows, columns and channels like:
    
(rows, columns, channels)

```python
im.shape
```

We will now create a dataframe with the co-ordinates of the places in Wuhan city where the cases are recorded for covid-19

```python
data = pd.DataFrame({'x': [230,254,440,250,331,500], 'y': [130,430,334,555,390,700]})
data.head()
```

Now we start to re-size the imag e to required output expectations. Now this step will change the input image size of (600,800) to (650,800).
This can be achieved using .resize() feature available in cv2

```python
im1 = cv2.resize(im, (650,800))
```

```python
plt.imshow(im1)
plt.show()
```

```python
data.head()
```

This method will show you how to create a object for plot using matplotlib as "plt.figure()". Then we will leverage this object to plot the scatter visual on top of base image of Wuhan city 

```python
fig = plt.figure()
plt.scatter(x=list(data['x']), y=list(data['y']), c='b', s=40)
```

```python
output = plt.imshow(im1)
plt.show()
```

Finally we will output the image as img format as double plot (image of Wuhan city + scatter plot of covid cases). The dpi paramter means the dots per inch which is the number of pixels per imch to imporve pixel quality

```python
fig.savefig('covid-cases.png',dpi=1200)
```
